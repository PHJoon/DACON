```python
import pandas as pd
import os
import shutil
import time
```

```python
train_df = pd.read_csv("/content/drive/MyDrive/efficientnet/전인서/train_df.csv")
label_list = train_df["label"].unique().tolist()
label_list = sorted(label_list)
len(label_list)  # type = list
```

```python
label_list
```

```python
# 폴더 안 파일 개수를 가져오는 함수
def get_num_of_files(directory):
  dirListing = os.listdir(directory)
  return len(dirListing)


# 15개 good 폴더의 파일 개수만 출력해보기
for label in label_list:
  num_files = get_num_of_files(f'/content/drive/MyDrive/efficientnet/전인서/train/{label}/')
  if num_files > 50:
    print(num_files)
```

### -> good 중의 최솟값인 60에 맞춰서 이상치를 오버샘플링해준다.

```python
# 이상치 폴더 이름 리스트
anomaly_label_list = []

for label in label_list:
  if label[-4:] != 'good':
    anomaly_label_list.append(label)


len(anomaly_label_list) # 88-15
```

```python
# 이상치 폴더 73개에 대해 for loop
# 폴더 내 파일들을 복사하되, 폴더 내 파일 개수가 60개 이상이 되면 stop

for dir in anomaly_label_list:
  current_dir = f'/content/drive/MyDrive/efficientnet/전인서/train/{dir}'
  file_list = os.listdir(current_dir)
  while len(file_list) < 60:
    for file in file_list:
      t = float(time.time())
      shutil.copyfile(f'{current_dir}/{file}', f'{current_dir}/{file}-{t}')
    file_list = os.listdir(current_dir)
  print(f'{dir} complete')
```
